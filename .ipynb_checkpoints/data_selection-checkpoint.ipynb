Load the cleaned data

```python
tmdb_df = pd.read_csv('output/tmdb_5000_cleaned.csv')
tmdb_df.head()
```

```python
tmdb_df['release_date']=pd.to_datetime(tmdb_df['release_date'], format='%Y-%m-%d', errors='ignore')
```

```python
def to_df(s):
    return pd.DataFrame(json.loads(s))
```

### Column 3: "Cast"

```python
avatar_cast = tmdb_df.loc[0, 'cast'].copy()
avatar_cast
```

```python
def uncredited_cost(character):
    return '(uncredited)' in character
```

```python
def remove_uncredited(character):
    return character.replace(' (uncredited)','')
```

```python
avatar_cast['uncredited'] = avatar_cast['character'].apply(uncredited_cost)
avatar_cast['character'] = avatar_cast['character'].apply(remove_uncredited)

avatar_cast
```

```python
avatar_cast['uncredited'].value_counts()
```

```python
def rename_gender(gender):
    if gender == 1:
        return "F"
    elif gender == 2:
        return "M"
    else:
        return "N/A"
```

```python
avatar_cast['gender'] = avatar_cast['gender'].apply(rename_gender)
avatar_cast
```

```python
np.sum(n_cast<10), np.sum(n_cast<5), np.sum(n_cast<3)
```

```python
tmdb_df['cast_num'] = n_cast
```

```python
def main_cast(cast, n_main=3):
    cast_pattern = np.array([np.nan, None, np.nan])
    if cast.empty:
        return np.tile(cast_pattern, n_main)
    else:
        n_cast = len(cast)
        cast_extract = cast.loc[:min(n_main-1, n_cast), ['id','name', 'gender']].values.flatten()
        return np.append(cast_extract, np.tile(cast_pattern, max(0, n_main-n_cast)))
```

```python
def cast_gender(cast):
    gender_counts = np.full(3, 0)
    if not cast.empty:
        groupby_counts = cast['gender'].value_counts()
        
        for i in range(3):
            if i in groupby_counts.index:
                gender_counts[i]= groupby_counts[i]
    
    return gender_counts
```

```python
cast_gender(obama_cast)
```

```python
tmdb_df_cast_gender = tmdb_df['cast'].apply(cast_gender)
```

```python
cast_gender_columns = ['cast_gender_na', 'cast_gender_f', 'cast_gender_m']

cast_gender_df = pd.DataFrame(np.array(tmdb_df_cast_gender.values.tolist()), columns=cast_gender_columns)
cast_gender_df
```

```python
n_genres.describe()
```

```python
n_genres.value_counts().sort_index()
```

```python
def main_genres(genres, n_main=3):
    genres_pattern = np.array([np.nan, None])
    if genres.empty:
        return np.tile(genres_pattern, n_main)
    else:
        n_genres = len(genres)
        genres_extract = genres.loc[:min(n_main-1, n_genres), ['id','name']].values.flatten()
    
    return np.append(genres_extract, np.tile(genres_pattern, max(0, n_main-n_genres)))
```

```python
tmdb_df_main_genres = tmdb_df['genres'].apply(main_genres)
```

```python
main_genres_columns = [['genre_'+str(i)+'_id', 'genre_'+str(i)+'_name'] for i in range(3)]
main_genres_columns = list(chain.from_iterable(main_genres_columns))

main_genres_df = pd.DataFrame(np.array(tmdb_df_main_genres.values.tolist()), columns=main_genres_columns)
main_genres_df
```

```python
tmdb_df = tmdb_df.join(main_genres_df)
tmdb_df.head()
```
