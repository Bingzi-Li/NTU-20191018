## 1 Apriori Algorithm

```python
from apyori import apriori
```

```python
def dataframe_asso_rules(rules):
    matrix = []
    for rule in rules:
        row = []
        
        genres = list(rule[0])
        row.append(len(genres))
        
        if len(genres)==1:
            row.append(genres[0])
        else:
            row.append(genres[0] + " -> " + genres[1])

        row.append(str(rule[1]))
        row.append(str(rule[2][0][2]))
        
        matrix.append(row)
    
    return pd.DataFrame(matrix, columns=['items', 'rule', 'support', 'confidence'])
```

```python
def print_asso_rules(rules):
    for rule in rules:

        genres = list(rule[0])

        if len(genres)==1:
            print("Simple Rule: " + genres[0])
        else:
            print("Association Rule: " + genres[0] + " -> " + genres[1])

        print("Support: " + str(rule[1]))

        print("Confidence: " + str(rule[2][0][2]))
        print()
```

```python
genre_asso_rules = list(apriori(tmdb_genre, max_length=2))
```

```python
len(genre_asso_rules)
```

```python
dataframe_asso_rules(genre_asso_rules)
```
